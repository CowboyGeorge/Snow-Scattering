```python
%matplotlib inline
import matplotlib.pyplot as plt
from pytmatrix.tmatrix import Scatterer
from pytmatrix.psd import PSDIntegrator, GammaPSD, BinnedPSD
from pytmatrix import orientation, radar, tmatrix_aux, refractive
import numpy as np

Rayleigh = []
flec1, flec2, flec3, flec4, flec5 = [], [], [], [], []
ar = 0.99999
#dees = np.logspace(-1, 1, 500)
dees = 1000*np.array([7.50e-05, 0.00015, 0.000225, 0.000275, 0.000325, 0.000375, 0.0004375, 0.0005125, 0.0005875, 0.0006625,
                 0.00075, 0.00085, 0.00095, 0.0011, 0.0013, 0.0015, 0.0017, 0.002, 0.0024, 0.0028, 0.0032, 0.0036,
                 0.004, 0.0044, 0.0048, 0.0055, 0.0065, 0.0075, 0.0085, 0.0095, 0.011, 0.013, 0.015, 0.017, 0.019])

ars = [0.5, 0.6, 0.7, 1.]


def mix(a, b, d, ar):
    '''
    I use the bruggeman mixing equation, at low densities of snowfall the decision between mixing
    equations is trivial. a and d are in cgs.
    '''
    m = a*d**b
    v = (1./6.)*ar*(d)**3*np.pi
    density = m/v
    n = 1.782
    kappa = complex(0, 7.302*10**-3)
    IceDens = .917
    e1 = n**2 - kappa**2
    e2 = 2*n*kappa
    B = e1 + e2
    f = density/IceDens
    mixed = (B*(1+2*f)-(2*f-2))/((2+f)+B*(1-f))
    e1out = np.real(mixed)
    e2out = np.imag(mixed)
    nout = np.sqrt((np.sqrt(e1out**2 +e2out**2)+e1out)/2)
    kout = np.sqrt((np.sqrt(e1out**2 +e2out**2)-e1out)/2)
    m =complex(nout, kout)
    return m

'''
Loop to generate cross sections
'''
for d in dees:
    print(d)
    dmod5 = d * (ar**(1./3.))
    dmod6 = d * (ar**(1./3.))
    dmod7 = d * (ar**(1./3.))
    dmod1 = d * (ar**(1./3.))
    mm = mix(0.004, 2.1, d*0.1, ar)
    
    oblate1 = Scatterer(radius = dmod5/2., wavelength=tmatrix_aux.wl_Ku, m = mm, axis_ratio = 1/ar)
    oblate2 = Scatterer(radius = dmod6/2., wavelength=tmatrix_aux.wl_Ku, m = mm, axis_ratio = 1/ar)
    oblate3 = Scatterer(radius = dmod7/2., wavelength=tmatrix_aux.wl_Ku, m = mm, axis_ratio = 1/ar)
    oblate4 = Scatterer(radius = dmod1/2., wavelength=tmatrix_aux.wl_Ku, m = mm, axis_ratio = 1/ar)
    
    flec1.append(10*np.log10(radar.refl(oblate1)))
    flec2.append(10*np.log10(radar.refl(oblate2)))
    flec3.append(10*np.log10(radar.refl(oblate3)))
    flec4.append(10*np.log10(radar.refl(oblate4)))

```
